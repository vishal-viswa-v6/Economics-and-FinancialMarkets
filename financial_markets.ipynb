#### At it's core, a financial market is a coordination mechanism, it matches buyers and sellers of financial instruments in order to allocate capital efficiently. Coordination mechanism fundamentally transfers capital from savers, who want returns to borrowers who need funding, discover prices for assets through supply and demand analysis, provide liquidity so that assets can be bought and sold efficiently and allow risk to be managed or transgered via derivatives.

#### Primary markets are where new securities are issued and secondary markets are where existing securities are traded. Centralized exchanges falcilitate all trades with high liquidity and counterparty risk but are prone to outages, regulation and manupilation. Decentralized exchanges do peer to peer trading using smart contracts which are trustless, transparent but faces slippage, MEV and gas fees. Direct trades without an exchange are over the counter, large orders will face lower slippage but opaque pricing. 

#### Financial instruments represent equities, bonds, commodities, currencies and cryptocurrencies. Equities/Stocks consist of ownership shares in a company, which entitles holders to profits and potential capital gains. Bonds are debt instruments, where investors lend money to issuers in exchange for periodic interest and return of principal. Commodities are raw materials like gold and wheat, traded as futures or EFTs. Currencies are trading pairs of global fiat currencies with high liquidity and are macro-driven, dominated by bank policy, inflation expectations and relative economic strength. Cryptocurrencies like Bitcoin (BTC) and Ethereum (ETH) are base layer assets with monterary properties and network utility, BTC stores of value and ETH powers smart contracts. Stablecoins, pegged to fiat, used for on-chain liquidity, trading pairs and yeild farming. DeFI tokens are utility in decentralized protocols, priced via useage, tokenonmics and speculation. NFTs are non fungible, illiquid tokens representing ownership of digital media or rights. Volatilty is extemely high, NFTs are speculative and niche. 




### Risk and Return 

#### Key types of risk include volatility, market risk, liquidity risk, credit risk, counterparty risk and Risk-Return/Trade-Off. Volatility is the most direct measure defined as the standard deviation of returns, it reflects how much an asset's price fluctuates over time, and is central to all quant models. Market risk, or systematic risk, affects all assets broadly and cannot be diversified away, examples include recessions or interest rate shocks. Liquidity risk arises when an asset can't be sold quickly without causing a large price impact this is especially relevant in crypto, where thin order books or unstable markets can trap capital. Credit risk refers to the chance of a borrower defaulting on obligations, relevant for bonds or lending platforms like Aave. Counterparty risk is the possibility that the other side of a trade (an exchange, broker, or DeFi protocol) fails to meet their obligation, this includes events like centralized exchange collapses or smart contract exploits. Tying all of this together is the risk-return trade-off, the core idea that higher potential returns come with greater risk exposure. T-bills offer low risk and low return; on the other end, something like a leveraged DeFi yield farm promises high returns but exposes capital to extreme downside.

```python
import matplotlib.pyplot as plt
import numpy as np

crypto_assets = {'BTC': [50, 60], 'ETH': [70, 80], 'SOL': [100, 120], 'USDC': [5, 1]}

returns = [v[0] for v in crypto_assets.values()]
volatility = [v[1] for v in crypto_assets.values()]
labels = list(crypto_assets.keys())

plt.figure(figsize=(10, 6))
plt.scatter(volatility, returns, color='teal')

for i, label in enumerate(labels):
    plt.annotate(label, (volatility[i]+1, returns[i]+1), fontsize=9)
    
plt.title('Crypto Market: Risk vs Return')
plt.xlabel('Volatility (Standard Deviation %)')
plt.ylabel('Annualized Return %')
plt.grid(True)
plt.tight_layout()
plt.show()
```
