### Gross Domestic Product (GDP)

#### GDP is the total market value of all final goods and services produced within a country's borders in a given period. Final goods refers to products sold to end users, if Tesla sells a car, that is counted, if a tire company sells tired to Tesla, that is not counted, it is embedded in the final product. 

#### Nominal GDP is not adjusted for inflation, it just sums up current prices in a given year.

#### Real GDP is adjusted for inflation. It reflects the actual growth in output, not price increases.

#### Example: If India's GDP grew from **\$2,000,000,000,000** to **\$2,200,000,000,000**, but inflation was 10%, then the Real GDP has not grown, the prices have gone up. 
#### Calculating GDP: 
$
\text{Real GDP} = \frac{\text{Nominal GDP}}{\text{GDP Deflator}} \times 100
$
#### GDP Deflator messures the price level of all new, domestically produced goods and services in an economy. 
#### Calculating GDP Deflator:
$
\text{GDP Deflator} = \frac{\text{Nominal GDP}}{\text{Real GDP}} \times 100
$
#### Growth Rate of GDP is the percentage increased in real GDP from one period to the next, this is the raw signal of economic growth. 
#### Calculate growth rate:
$
\text{GDP Growth Rate} = \frac{\text{Real GDP}_{\text{New}} - \text{Real GDP}_{\text{Old}}}{\text{Real GDP}_{\text{Old}}} \times 100
$
#### If real GDP grew from **\$1.5T** to **\$1.65T** = $ (1.65 - 1.5)/1.5 = 10 $% Growth 

#### Uncaptured GDP includes black market trades, unpaid work, externalities as in damage done to increase GDP not calculated, income distribution within the context of inequality, quaility improvements such as a product with the same nomial price but higher quality from improvements over time.

### GDP vs GNP (Gross National Product)
#### GDP is domestic, counts everything produced inside a country, including foreign companies operating in the same country. Toyota factory in the United States counts as U.S GDP not Japan's GDP. 
#### GNP is national, counts everything produced by a country's citizens and companies, irrelavent of the location. Infosys earnings from projects in the United States counts as India's GNP (Gross National Product) and not U.S GNP.

#### GNP = GDP + Net Income Abroad 

```python
years = np.arange(2010, 2025)
gdp = 15000 * (1.02 ** np.arange(15))
np.random.seed(0)
net_income_abroad = np.random.randint(80, 160, size = 15)

gnp = gdp + net_income_abroad

plt.figure(figsize=(10, 6))
plt.plot(years, gdp, label = "GDP", marker = "o")
plt.plot(years, gnp, label = "GNP", marker = "s")

plt.title('GDP vs GNP over time')
plt.xlabel('year')
plt.ylabel('Value in Billions USD')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Inflation

#### Inflation is the rate at which the general level of prices for goods and services rise over time. When inflation occurs, each unit of currency buys fewer goods and services. Inflation happens because of imbalances between supply and demand, rising costs and policy decisions. When demand exceeds supply, too much money is spent chasing too few goods. When costs rise, producers pass those increases to consumers. Policy choices in central banks and/or governments may create inflation due to printing money or spending programs. 

#### How does inflation happen?
#### Common causes are, consumers spending more, businesses cannot keep up with demand, prices go up as businesses raise prices to balance supply and demand, key input costs such as oil and raw materials increase in prices, production costs increase, final prices increase to maintain margins, workers demand higher pay, businesses rise prices to cover higher pay.  

#### Consumer Price Index measures the average change in prices paid by consumers for a fixed basket of goods and services over time. Producer Price Index measures the average change in selling prices received by domestic producers for their output. GDP Deflator measures the change in prices for all goods and services included in GDP.  

#### Nominal value is the actual monetary value at the time. 
#### Real value is the inflation adjusted value showing purchasing power:
$\text{Real Value} = \frac{\text{Nominal Value}}{\text{Price Index}} \times 100$

### Unemployment


#### Unemployment is a share of the labor force that is willing and able to work but cannot find a job.

#### Unemployment occurs when there is lack of demand for labor, mismatch of skillsets required, workers are in between jobs or entering the workforce. Structural unemployment is caused by a mismatch between a workers skills and job requirements. Frictional unemployment is people moving between jobs or entering the workforce. Cyclical unemployment is due to economic downturns, lack of demand for goods and services, businesses lay off workers. Natural unemployment exists due to people switching jobs or retraining, so unemployment is never zero.

### Aggregate Demand (AD) and Aggregate Supply (AS)

#### AD is the total quantity of goods and services demanded in an economy at various price levels. As price levels fail, real wealth increases, interest rates fall, exports become more competitve therefore demand increases.

#### AD = C + I + G + (X - M) where C = Consumption, I = Investment, G = Gov Spending, X - M = Net Exports

#### AS shows the total output producers are willing to supply at different price levels. Short-run AS is upward sloping, firms respond to higher prices with more production. Long-run AS is vertical, economy is at full capacity, prices rise and output does not.
